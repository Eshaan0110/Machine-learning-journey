```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create dataset (10 rows, 3 parameters)
data = {
    "Hours_Studied": [2, 4, 6, 8, 10, 3, 7, 5, 9, 1],
    "Projects_Completed": [1, 2, 3, 4, 5, 2, 4, 3, 5, 1],
    "Internships": [0, 1, 1, 2, 2, 0, 2, 1, 2, 0],
    "Salary": [20, 35, 50, 65, 80, 30, 70, 55, 75, 15]
}

df = pd.DataFrame(data)
df
```

```python
# Independent variables
X = df[["Hours_Studied", "Projects_Completed", "Internships"]].values

# Add intercept column (1s)
X = np.c_[np.ones(len(X)), X]

# Dependent variable
y = df["Salary"].values.reshape(-1,1)

X, y
```

```python
# Normal equation
beta = np.linalg.inv(X.T @ X) @ X.T @ y

beta
```

```python
print("Regressor Equation:")
print(f"Salary = {beta[0][0]:.3f} + {beta[1][0]:.3f}(Hours) + {beta[2][0]:.3f}(Projects) + {beta[3][0]:.3f}(Internships)")
```

```python
y_pred = X @ beta
y_pred
```

```python
n = len(y)

# Mean Squared Error
mse = np.sum((y - y_pred)**2) / n

# Root MSE
rmse = np.sqrt(mse)

# Mean Absolute Error
mae = np.sum(np.abs(y - y_pred)) / n

# R-squared
ss_total = np.sum((y - np.mean(y))**2)
ss_residual = np.sum((y - y_pred)**2)

r2 = 1 - (ss_residual / ss_total)

print("MSE:", mse)
print("RMSE:", rmse)
print("MAE:", mae)
print("R2 Score:", r2)
```

```python
new_x = np.array([[1, 6, 3, 1]])
predicted_salary = new_x @ beta

print("Predicted Salary (Multiple Regression):", predicted_salary[0][0])
```

```python
X_poly2 = df[["Hours_Studied", "Projects_Completed", "Internships"]].values

h = X_poly2[:,0]
p = X_poly2[:,1]
i = X_poly2[:,2]

X2 = np.column_stack([
    np.ones(len(h)),
    h, p, i,
    h**2, p**2, i**2,
    h*p, h*i, p*i
])

y = df["Salary"].values.reshape(-1,1)

X2.shape
```

```python
beta2 = np.linalg.inv(X2.T @ X2) @ X2.T @ y
beta2
```

```python
y_pred2 = X2 @ beta2

mse2 = np.sum((y - y_pred2)**2) / len(y)
rmse2 = np.sqrt(mse2)
mae2 = np.sum(np.abs(y - y_pred2)) / len(y)

ss_total2 = np.sum((y - np.mean(y))**2)
ss_residual2 = np.sum((y - y_pred2)**2)

r2_2 = 1 - (ss_residual2 / ss_total2)

print("Degree 2 Metrics")
print("MSE:", mse2)
print("RMSE:", rmse2)
print("MAE:", mae2)
print("R2:", r2_2)
```

```python
X3 = np.column_stack([
    np.ones(len(h)),
    h, p, i,
    h**2, p**2, i**2,
    h**3, p**3, i**3,
    h*p, h*i, p*i
])

X3.shape
```

```python
beta3 = np.linalg.inv(X3.T @ X3) @ X3.T @ y
beta3
```

```python
y_pred3 = X3 @ beta3

mse3 = np.sum((y - y_pred3)**2) / len(y)
rmse3 = np.sqrt(mse3)
mae3 = np.sum(np.abs(y - y_pred3)) / len(y)

ss_total3 = np.sum((y - np.mean(y))**2)
ss_residual3 = np.sum((y - y_pred3)**2)

r2_3 = 1 - (ss_residual3 / ss_total3)

print("Degree 3 Metrics")
print("MSE:", mse3)
print("RMSE:", rmse3)
print("MAE:", mae3)
print("R2:", r2_3)
```

```python
# New input
h_new = 6
p_new = 3
i_new = 1

# Degree 2 prediction
new_x2 = np.array([[1,
                    h_new, p_new, i_new,
                    h_new**2, p_new**2, i_new**2,
                    h_new*p_new, h_new*i_new, p_new*i_new]])

pred2 = new_x2 @ beta2

# Degree 3 prediction
new_x3 = np.array([[1,
                    h_new, p_new, i_new,
                    h_new**2, p_new**2, i_new**2,
                    h_new**3, p_new**3, i_new**3,
                    h_new*p_new, h_new*i_new, p_new*i_new]])

pred3 = new_x3 @ beta3

print("Predicted Salary (Degree 2):", pred2[0][0])
print("Predicted Salary (Degree 3):", pred3[0][0])
```
